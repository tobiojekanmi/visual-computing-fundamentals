```python
import matplotlib.pyplot as plt
import pathlib
from utils import read_im, save_im
output_dir = pathlib.Path("image_solutions")
output_dir.mkdir(exist_ok=True)
```

```python
im = read_im(pathlib.Path("images", "duck.jpeg"))
plt.imshow(im)
```

## Task 2a

```python
def greyscale(im):
    """ Converts an RGB image to greyscale
    
    Args:
        im ([type]): [np.array of shape [H, W, 3]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """

    im = (0.212 * im[:,:, 0]) + (0.7152 * im[:,:, 1]) + (0.0722 * im[:,:, 2])

    return im
```

```python
im_greyscale = greyscale(im)
save_im(output_dir.joinpath("duck_greyscale.jpeg"), im_greyscale, cmap="gray")
plt.imshow(im_greyscale, cmap="gray")
```

## Task 2b

```python
def inverse(im):
    """ Finds the inverse of the greyscale image
    
    Args:
        im ([type]): [np.array of shape [H, W]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """    
    # YOUR CODE HERE
    if im.max() > (1.0 + 1e-3):
       im = 255.0 - im
    else:
       im = 1.0 - im

    return im
```

```python
im_inv_greyscale = inverse(greyscale(im))
save_im(output_dir.joinpath("duck_inv_greyscale.jpeg"), im_inv_greyscale, cmap="gray")
plt.imshow(im_inv_greyscale, cmap="gray")
```
